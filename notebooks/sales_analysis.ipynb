# Beauty Product Sales Analysis

This project analyzes sales data for beauty products to identify top-performing items based on total sales.

## Key Steps:
- Load and explore the dataset
- Group sales by product
- Identify top-performing products
- Visualize results using a bar chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('sales_data.csv')

print(df.head())
```

```python
top_products = df.groupby('Product')['Sales'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8, 5))

ax = sns.barplot(
    x=top_products.values,
    y=top_products.index,
    hue=top_products.index,
    palette='viridis',
    legend=False
)

plt.title('Top Beauty Products by Sales')
plt.xlabel('Total Sales ($)')
plt.ylabel('Product')

for i, value in enumerate(top_products.values):
    ax.text(value + 0.3, i, str(value), va='center')

plt.tight_layout()
plt.show()
```
